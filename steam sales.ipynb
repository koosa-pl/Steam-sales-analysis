```python
x = ''
df = df[df['reviews'] != x] # deleting games from the df with no reviews
```

```python
df.loc[df['reviews'] == "Mostly Positive"]
```

```python
df = df[df['reviews'] != "Mostly Positive"]  # deleting one outlier with incorrect value
```

```python
df['reviews'] = df['reviews'].astype(int) # setting reviews columns as an integer type for sorting
```

```python
df.dtypes
```

```python
df = df.sort_values(by=['reviews'], ascending=False) # sorting steam games by number of reviews
```

```python
df2 = df.iloc[:1000] # grabbing top 1000 most popular steam games by number of reviews
```

```python
df2
```

```python
df2['genres'].value_counts()
```

```python
genre_list = df2['genres'].unique()
```

```python
genres_simplified = []

for genre in genre_list:
    genre2= genre.split(";")
    if len(genre2) > 1:
        multigenre = f'{genre2[0]}/{genre2[1]}'
        genres_simplified.append(multigenre)
    else:
        genres_simplified.append(genre2[0])

print(genres_simplified) # simplifying number of possible genre combinations
```

```python
len(genre_list)
```

```python
len(genres_simplified)
```

```python
len(set(genres_simplified)) # number of game genres reduced from 132 to 46
```

```python
pd.options.mode.chained_assignment = None # setting needed, otherwise code below would result in a warning
```

```python
df2['genres'] = df2['genres'].replace(df2['genres'].unique(), genres_simplified)
```

```python
test = df2['genres'].unique()
set(test) == set(genres_simplified) 
# checking if values from Dataframe were replaced correctly by matching them with simplified genre list
```

```python
mp_or_sp = df2['multiplayer_or_singleplayer'].unique()
```

```python
no_of_players_simplified = []

for x in mp_or_sp:
    mp_or_sp2 = x.split(';')
    no_of_players_simplified.append(mp_or_sp2[0])  # simplifying the single/multiplayer category
```

```python
df2['multiplayer_or_singleplayer'] = df2['multiplayer_or_singleplayer'].replace(df2['multiplayer_or_singleplayer'].unique(),no_of_players_simplified)
```

```python
df2.reset_index(inplace=True)
df2.drop(['id'], axis=1, inplace=True) # index reset and dropping id column which is not needed
```
